Primero cargamos las librerías y los datos necesarios.

```python
import pandas as pd
import numpy as np

# Cargar desde CSV
df_train = pd.read_csv('df_train.csv')
df_test = pd.read_csv('df_test.csv')
```

2. Análisis de ciclos máximos realizados por los motores.


```python
import matplotlib.pyplot as plt
import seaborn as sns

maxCycles = []

#Inicialización de variables

max_time = [] #Array con el numero


#Extracción de array de maximos ciclos por motor.
for unit_number in df_test['unit_number'].unique():
  df_max_cycles = df_test[df_test['unit_number'] == unit_number]
  mx_time = df_max_cycles['time'].max()
  max_time.append(mx_time)

#Variable global array 2D para guardar los ciclos maximos por experimento
maxCycles.append(max_time)
```

Histogramas con frecuencias de ciclos por motor

```python
for max_time in maxCycles:
   # Variables descriptivas
    mean_max_time = np.mean(max_time) #Promedio de ciclos completados por un motor
    max_value = np.max(max_time) #Ciclos maximos completaods por un motor
    min_value = np.min(max_time) #Ciclos minimos completaods por un motor

    print(f"Promedio de cilos:", mean_max_time)
    print(f"Valor maximo de ciclos:", max_value)
    print(f"Valor minimo de ciclos: ", min_value)

    #Presentación del histograma.
    sns.set_style("white")
    plt.figure(figsize=(5, 3))
    label = str('Ciclos por motor')
    sns.histplot(max_time, bins=10, kde=True, color='k', label= label, alpha=0.5,
                line_kws={'color': 'orange', 'lw': 2, 'ls': ':'})
    plt.xlabel('Número de ciclos máximos por motor')
    plt.ylabel('Frecuencia')
    plt.title('Histograma de ciclos máximos por motor entre todas las condiciones de prueba')
    plt.legend()
    plt.grid()
    name = label + ".png"
    plt.savefig(name, dpi=300, bbox_inches='tight')
    plt.show()
```

3. Mapa de correlación entre los datos

```python

#Definimos las correlaciones con in-built functions

df_experiment = df_test.drop('dataSet', axis = 1)
df_correlations = df_experiment.corr()

threshold = 0.98

mask = df_correlations.where ( (abs(df_correlations) >= threshold)).isna()

#settings de la figura
sns.set_style("whitegrid", {"axes.facecolor": ".0"})
plt.figure(figsize=(9,8))
plot_kws={"s": 1}

#Heat nao de correación -1 y 1 son correlaciones fuertes.
sns.heatmap(df_correlations,
  cmap='Spectral',
  annot = False,
  mask=mask,
  linewidths=0.1,
  linecolor='lightgrey').set_facecolor('white')
name =  'Mapa_de_correlacion.png'
plt.savefig(name, dpi=300)
plt.show()
```

4. Correlación de los sensores y las mediciones.

A partir de los mapas de correlación, y profundizar en la naturaleza de las variables, exploramos las distintas correlaciones para algunas variables que observamos, son las que mas correlaciones presentan.

Además de algunas variables que pareciera tienen correlaciones relevantes,en realidad no encontramos una causalidad entre ellas, por lo que en el análisis de correlación decidimos ignorarlas.

```python
import matplotlib.pyplot as plt

#Correlaciones para el sensor 6.

columnas = [
    ['senMes6', 'opSet1'],
    ['senMes6', 'senMes1'],
    ['senMes6', 'senMes5'],
    ['senMes6', 'senMes7'],
    ['senMes6', 'senMes12'],
    ['senMes6', 'senMes21'],
]

fig, axs = plt.subplots(2, 3, figsize=(20, 10))

for i, cols in enumerate(columnas):
    x = df_train[cols[0]]
    y = df_train[cols[1]]

    row = i // 3
    col = i % 3

    axs[row, col].scatter(x, y, marker='.', color='blue')
    axs[row, col].set_title(f'Scatter Plot {cols[0]} vs {cols[1]}')
    axs[row, col].set_xlabel(cols[0])
    axs[row, col].set_ylabel(cols[1])
    #axs[row, col].grid()

fig.patch.set_facecolor('white')
#plt.tight_layout()

plt.savefig('scatter_plots_senMes6.png')

plt.show()
```
